# Questão 01 <a name="1"></a>

Escreva um algoritmo que mostre na tela uma escada de tamanho n utilizando o caractere * e espaços. A base e altura da escada devem ser iguais ao valor de n. A última linha não deve conter nenhum espaço.


```python
# A funcao escada conta quantos degraus se deseja, pelo loop ela conta para os espacos dum lado e asteriscos do outro #

def escada(n_degraus):
    for degrau in range(1, n_degraus + 1):
        print(' ' * (n_degraus - degrau) + '*' * degrau)
        

escada(int(input("Digitar numero de degraus (n): ")))
```

# Questão 02 <a name="2"></a>

Débora se inscreveu em uma rede social para se manter em contato com seus amigos. A página de cadastro exigia o preenchimento dos campos de nome e senha, porém a senha precisa ser forte. O site considera uma senha forte quando ela satisfaz os seguintes critérios:
•	Possui no mínimo 6 caracteres.
•	Contém no mínimo 1 digito.
•	Contém no mínimo 1 letra em minúsculo.
•	Contém no mínimo 1 letra em maiúsculo.
•	Contém no mínimo 1 caractere especial. Os caracteres especiais são: !@#$%^&*()-+
Débora digitou uma string aleatória no campo de senha, porém ela não tem certeza se é uma senha forte. Para ajudar Débora, construa um algoritmo que informe qual é o número mínimo de caracteres que devem ser adicionados para uma string qualquer ser considerada segura.


```python
# A funcao criador de senha testa se a senha desejada atende as condicionais informadas#
def criador_de_senha(senha):
    
    caracteres_especiais = ['!', '@', '#', '$', '%', '^', '&', '*', '()', '-', '+']
    cond = True
    
    if len(senha) < 6:
        print('Faltam', str(6 - len(senha)), 'caracteres!', '\nSenha deve ter no minimo 6 caracteres!')
        cond = False
        
    if not any(caracter.isdigit() for caracter in senha):
        print('Senha deve ter no minimo um numero! ')
        cond = False
        
    if not any(caracter.islower() for caracter in senha):
        print(('Senha deve ter no minimo uma letra minuscula! '))
        cond = False
        
    if not any(caracter.isupper() for caracter in senha):
        print(('Senha deve ter no minimo uma letra maiuscula! '))
        cond = False
        
    if not any(caracter in caracteres_especiais for caracter in senha):
        print(('Senha deve ter no minimo um caracter especial! '))
        cond = False
        
    if cond:
        return cond
      
criador_de_senha(input())
```

# Questão 03 <a name="3"></a>
Duas palavras podem ser consideradas anagramas de si mesmas se as letras de uma palavra podem ser realocadas para formar a outra palavra. Dada uma string qualquer, desenvolva um algoritmo que encontre o número de pares de substrings que são anagramas.


```python
def all_perms(elements):
    if len(elements) <=1:
        return elements
    else:
        tmp = []
        for perm in all_perms(elements[1:]):
            for i in range(len(elements)):
                tmp.append(perm[:i] + elements[0:1] + perm[i:])
        tmp = set(tmp)
                
    return tmp

def contagem(n):
    return len(n)     

contagem(all_perms(input()))
```
